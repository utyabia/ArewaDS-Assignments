```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('../../data/birds.csv')
```

```python
order_name = df.loc[df['Name'] == "Fulvous whistling-duck", 'Order'].values[0]
order_birds = df[df['Order'] == order_name]

#conservation status

status_counts = order_birds['ConservationStatus'].value_counts()
status_counts
```

```python
status = df.loc[df['Name'] == "Snow goose", 'ConservationStatus'].values[0]
print("Snow Goose Conservation Status:", status)
```

```python
plt.figure(figsize=(8,5))
status_counts.plot(kind='bar')
plt.title(f"Conservation Status of Birds in Order: {order_name}")
plt.xlabel("Conservation Status")
plt.ylabel("Number of Species")
plt.show()
```

According to the dataset, Snow Geese are classified as LC (Least Concern). This means they are not currently at risk of extinction, unlike some of their relatives in the Anatidae family.

```python
snow_goose = df[df['Name'] == "Snow goose"].iloc[0]

lengths = [snow_goose['MinLength'], snow_goose['MaxLength']]

# Plot line from min to max
plt.figure(figsize=(6,4))
plt.plot(["Min Length", "Max Length"], lengths, marker="o")
plt.title("Snow Goose Length Range")
plt.ylabel("Length (cm)")
plt.show()
```

```python
family_name = snow_goose['Family']
family_birds = df[df['Family'] == family_name]

plt.figure(figsize=(8,6))
plt.scatter(family_birds['MaxWingspan'], family_birds['MaxBodyMass'], alpha=0.6, label="Other Family Members")

# Highlight Snow Goose
plt.scatter(snow_goose['MaxWingspan'], snow_goose['MaxBodyMass'],
            color='red', s=100, label="Snow goose")

plt.title(f"Wingspan vs Body Mass in Family: {family_name}")
plt.xlabel("Max Wingspan (cm)")
plt.ylabel("Max Body Mass (g)")
plt.legend()
plt.show()
```


This scatterplot compares the maximum wingspan and body mass of all birds in the Snow Gooses family (Anatidae). The Snow Goose (highlighted in red) is one of the heavier members, though its wingspan is not the largest  showing that it is more robust than wide-winged compared to its relatives.
