```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Titanic dataset from pandas
df = sns.load_dataset("titanic")

# Preview the first rows
print(df.head())
```

```python
plt.figure(figsize=(7,5))
plt.hist(df['age'].dropna(), bins=30, edgecolor='black')
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

Based on the above plot, most passengers were between 20–40 years old, showing the Titanic carried many young adults. Very few elderly passengers were onboard.

```python
plt.figure(figsize=(7,5))
plt.hist(df['fare'], bins=40, edgecolor='black')
plt.title("Distribution of Ticket Fares")
plt.xlabel("Fare ($)")
plt.ylabel("Number of Passengers")
plt.show()
```

The fare distribution is right-skewed. Most passengers paid relatively low fares, while a small number of very wealthy passengers paid extremely high ticket prices.

```python
survived = df[df['survived'] == 1]['age'].dropna()
not_survived = df[df['survived'] == 0]['age'].dropna()

plt.figure(figsize=(7,5))
plt.hist([survived, not_survived], bins=30, stacked=True,
         label=['Survived', 'Did Not Survive'], edgecolor='black')
plt.title("Age Distribution by Survival")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()
```

This histogram shows that children had higher survival rates compared to adults — reflecting the evacuation policy of “women and children first.” Adults had lower chances of survival, especially in the middle-aged groups.
